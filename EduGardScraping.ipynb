```
import requests
import urllib
from bs4 import BeautifulSoup
from IPython.display import Image,display

url = 'https://www.mi.com/in/list/'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

phone_div = soup.find('div',{"class":"box-bd J_box-bd"})

#soup finds out all "a" tag having image link from the div with class "box-bd J_box-bd" 
#and loads all these "a" tags in "img"
img = phone_div.find_all('a',{"class":"category-list-img"})

#same for storing all "a" tags having name of smartphones
name = phone_div.find_all('a',{"class":"category-list-title"})

# To Store properly formatted phone names by excluding HTML using .text and removing spaces using .strip()
phone_name=[]
for i in name:
    phone_name.append((i.text).strip())

# To Store all image links i.e. "src" from "img" under "a" tag
image_url=[]    
for image in img:
    link = image.find('img').get('src')
    image_url.append(link)
```

**Downloading and Displaying**




```
for i in range(0,20):
    name = phone_name[i]
    full_file_name = name+ '.jpg'
    link = image_url[i]
        
        # this downloads image from the link scraped and saves it with above full file name
    urllib.request.urlretrieve(link,full_file_name)
        
        # To display image and its name in jupyter after scraping and downloading data (image and name)
    display(Image(filename=full_file_name))
    print(name)
```

```
from google_images_download import google_images_download  
  
# creating object 
response = google_images_download.googleimagesdownload()  
  

def downloadimages(): 
    
    query = input('Enter Keyword for image')
    # keywords is the search query 
    # format is the image file format (.jpg, .png, .jpeg )
    # limit is the number of images you want to download 
    # print urls is to print the image file url while function executes
    # size is the image size which can  be ("large, medium, icon") 
    # aspect ratio denotes the height-width ratio of images to download, can be  ("tall, square, wide, panoramic") 

    arguments = {"keywords": query, 
                 "format": "jpg", 
                 "limit":5, 
                 "print_urls":True, 
                 "size": "large", 
                 "aspect_ratio": "panoramic"} 
    try: 
        response.download(arguments)
    except FileNotFoundError:
        print('File not found')
```

```
# Calling Function
downloadimages()
```
